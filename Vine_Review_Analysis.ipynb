```python
# Add the Pandas dependency.
import pandas as pd
```

```python
vine_data_to_load = "vine_table.csv"
vine_data_df = pd.read_csv(vine_data_to_load)
vine_data_df
```

```python
filtered_vine_df = vine_data_df[vine_data_df['total_votes']>=20]
filtered_vine_df
```

```python
# retrieve all the rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%.
helpful_vine_df = filtered_vine_df[(filtered_vine_df['helpful_votes'])/(filtered_vine_df['total_votes'])>=.5]
helpful_vine_df
```

```python
# retrieves all the rows where a review was written as part of the Vine program (paid), vine == 'Y'
paid_vine_df = helpful_vine_df[helpful_vine_df['vine']=='Y']
paid_vine_df
```

```python
# retrieve all the rows where the review was not part of the Vine program (unpaid), vine == 'N'.

unpaid_vine_df = helpful_vine_df[helpful_vine_df['vine']=='N']
unpaid_vine_df
```

# Determine the total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews for the two types of review (paid vs unpaid).

```python
paid_vine_reviews = paid_vine_df['review_id'].count()
paid_vine_reviews
```

```python
paid_vine_reviews_5star = paid_vine_df[paid_vine_df['star_rating'] == 5]['star_rating'].count()
paid_vine_reviews_5star
```

```python
paid_5star_percentage = paid_vine_reviews_5star/paid_vine_reviews*100
paid_5star_percentage
```

```python
print(f'{paid_5star_percentage:2f}% of paid reviews are 5 star')
```

```python
unpaid_vine_reviews = unpaid_vine_df['review_id'].count()
unpaid_vine_reviews
```

```python
unpaid_vine_reviews_5star = unpaid_vine_df[unpaid_vine_df['star_rating'] == 5]['star_rating'].count()
unpaid_vine_reviews_5star
```

```python
unpaid_5star_percentage = unpaid_vine_reviews_5star/unpaid_vine_reviews*100
unpaid_5star_percentage
```

```python
print(f'{unpaid_5star_percentage:2f}% of unpaid reviews are 5 star')
```
